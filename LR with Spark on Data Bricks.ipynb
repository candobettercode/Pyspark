```
#FIle Location  and type
file_location = "/FileStore/tables/tips.csv"
file_type = "csv"

df = spark.read.csv(file_location,header = True, inferSchema= True)
```

```
df.show()
```

```
df.printSchema()
```

```
df.columns
```

```
### Handling categorical features
from pyspark.ml.feature import StringIndexer
```

```
indexer=StringIndexer(inputCol='sex',outputCol='sex_indexed')
df_r=indexer.fit(df).transform(df)
df_r.show()
```

```
indexer=StringIndexer(inputCols=['sex','smoker','day','time'],outputCols=['sex_indexed','smoker_indexed','day_indexed','time_indexed'])
df_s=indexer.fit(df).transform(df)
df_s.show()
```

```
df_s.columns
```

```
from pyspark.ml.feature import VectorAssembler
featureassembler=VectorAssembler(inputCols=['tip','size','sex_indexed','smoker_indexed','day_indexed','time_indexed'], outputCol='Independet_Features')
output=featureassembler.transform(df_s)
```

```
output.show()
```

```
output.select('Independet_Features').show()
```

```
finalized_data=output.select('Independet_Features','total_bill')
```

```
finalized_data.show()
```

```
from pyspark.ml.regression import LinearRegression
## train test split
train_data,test_data = finalized_data.randomSplit([0.75,0.25])
regressor = LinearRegression(featuresCol='Independet_Features', labelCol='total_bill')
regressor=regressor.fit(train_data)
```

```
regressor.coefficients
```

```
regressor.intercept
```

```
###Prediction
pred_result=regressor.evaluate(test_data)
```

```
###Final Comparision
pred_result.predictions.show()
```

```
pred_result.r2,pred_result.meanAbsoluteError,pred_result.meanSquaredError
```
